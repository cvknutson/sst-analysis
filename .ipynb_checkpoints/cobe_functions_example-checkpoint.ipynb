# this code is designed to help demonstrate how to use the module code_functions

```python
# import the cobe_functions module
import numpy as np
import netCDF4 as nc
import cobe_functions as cf
import datetime

# read in your specific file
filename = "sst.mon.mean.nc"
```

From here on there will be specific functions and how to use them.

```python
# read_latest_sst
# this function is designed to give you the latest
# sst recorded in your netCDF file

sst_latest = cf.read_latest_sst(filename)
print(sst_latest)
```

```python
# read_sst
# this function is designed to convert your netCDF file into a dataset with
# variables of sst, lats, lons, and times

dataset = cf.read_sst(filename)
print(dataset)
```

```python
# load_time
# this function is designed to convert times in your netCDF file to an array of 
# year, month, day, hour, and minute

times = cf.load_time(filename)
print(times)
```
